```python
def pol_gen(coeff_vect):
    pl = 0
    d = len(coeff_vect)-1
    for i in range(0,len(coeff_vect)):
        pl = (pl + int(coeff_vect[d-i])*x**i)
    return pl

def all_coeff_gen(d):
    all_coeff = np.zeros((2**(d+1),d+1))
    for i in range(0,2**(d+1)): 
        t = i
        for j in range (0,d+1):
            all_coeff[i][d-j]= t%2
            t = int(t/2)
    return all_coeff

def mod2_pol(pol):
    if(pol==0):
        return pol_gen([0])
    else:
        d = degree(pol)
        coeff_vect = np.zeros((d+1,1))
        for i in range(0,d):
            qx,rx = div(pol,x**(d-i))
            coeff_vect[i] = qx%2
            pol = pol - qx*x**(d-i)
        coeff_vect[d] = pol%2
        return pol_gen(coeff_vect)
    
```

```python
# Finding all irreducible polynomial of degree 8 over F2 (Hit and trial)
d=8

# Generating all polynomials upto degree 7
all_coeff = all_coeff_gen(d-1)

print("All irreducible polynomial of degree 8 over F2:")
for i in range(0,2**d):
    flag = 0
    
    # Check for all possible monic polynomials of degree 8 whether they are irreducible or not 
    sx = x**d + pol_gen(all_coeff[i])

    # Check for any factors
    # Check for all polynomials upto degree 7 (8-1) 
    # 0,1 are not to be taken as factors that's why loop starts from 2 (1+x)        
    for j in range (2,2**d): 
        qx,rx= div(sx,pol_gen(all_coeff[j]))
        rx = mod2_pol(rx)  
        
        # If there exist any factor then the polynomial is reducible  
        if(rx==0):
            flag = 1
            break 
            
    # If there were no factors then the polynomial is irreducible 
    if(flag==0):
        print(sx)
```

## Define Functions 

```python
def gnrt_fnt_fld(d):
    F2_d = {}
    all_coeff = all_coeff_gen(d-1)
    for i in range(0,2**d):
        F2_d[i] = pol_gen(all_coeff[i])
    return F2_d

def fnd_irr_pol(d):
    cnt =0
    irr_pols = {}
    all_coeff = all_coeff_gen(d-1)
    for i in range(0,2**d):
        flag = 0
        sx = x**d + pol_gen(all_coeff[i])
        for j in range (2,2**d): 
            qx,rx= div(sx,pol_gen(all_coeff[j]))
            rx = mod2_pol(rx)  
            if(rx==0):
                flag = 1
                break 
        if(flag==0):
            irr_pols[cnt] = sx
            cnt = cnt + 1
    return irr_pols

def fnd_prim_el(F2_d,sx,d):
    prim_el = 0
    flag = 0
    for i in range(2,2**d):
        if (flag == 0):
            for j in range (1,2**d):
                qx,rx=div((F2_d[i])**(j),sx)
                rx = mod2_pol(rx)
                if (rx == 1):
                    if ( j!= 2**d-1):
                        break
                    elif (j==2**d-1):
                        prim_el = F2_d[i]
                        flag = 1
        else: 
            return prim_el
            break            
```

```python
def prim_el_rep(el,d,prim_el,sx):
    if (el == 0):
        return "z" 
    for i in range(0,2**d):
        qx,rx = div(prim_el**i,sx)
        temp = mod2_pol(rx)
        if (el == temp):
            return i
        
def prim_el_rep_add(p_1,p_2,d,prim_el,sx):
    if (p_1 !='z' and p_2!='z'):
        temp = prim_el**p_1 + prim_el**p_2
    elif (p_1 == 'z' and p_2!='z'):
        temp = 0 + prim_el**p_2
    elif (p_1 !='z' and p_2 == 'z'):
        temp = prim_el**p_1 + 0
    else:
        temp = 0         
    qx,rx = div(temp,sx)
    return prim_el_rep(mod2_pol(rx),d,prim_el,sx)
    
def prim_el_rep_sub(p_1,p_2,d,prim_el,sx):
    if (p_1 !='z' and p_2!='z'):
        temp = prim_el**p_1 - prim_el**p_2
    elif (p_1 == 'z' and p_2!='z'):
        temp = 0 - prim_el**p_2
    elif (p_1 !='z' and p_2 == 'z'):
        temp = prim_el**p_1 - 0
    else:
        temp = 0         
    qx,rx = div(temp,sx)
    return prim_el_rep(mod2_pol(rx),d,prim_el,sx)

def prim_el_rep_mul(p_1,p_2,d):
    if (p_1 == "z" or p_2 == "z"):
        return "z"
    else:
        return (p_1 + p_2)%(2**d-1)

def prim_el_rep_div(p_1,p_2,d):
    if (p_1 == "z" or p_2 == "z"):
        return "z"
    else:
        return (p_1 - p_2)%(2**d-1) 
    
def prim_el_rep_inverse(p):
    if (p=="z"):
        return "z"
    else:
        return (2**d-1-p)%(2**d-1)

def deg_ny_find(prim_el_rep_ny):
    degr = 0
    for i in range(len(prim_el_rep_ny)):
            if(prim_el_rep_ny[i]!="z"):
                degr = i
    return degr

def pol_from_prim_el_rep(prim_el_rep_vect,var):
    pol_var = 0
    for i in range(len(prim_el_rep_vect)):
        if (prim_el_rep_vect[i]=="z"):
            pol_var = pol_var + 0
        else: 
            pol_var = pol_var + prim_el_rep_vect[i]*var**i
    return pol_var
    

def pol_fq_div(ny,ey,d,prim_el,sx):
    cnt = 0
    prim_el_rep_ny = {}
    for i in range(len(ny)):
        prim_el_rep_ny[i] = prim_el_rep(ny[i],d,prim_el,sx)
    deg_ny = deg_ny_find(prim_el_rep_ny)
    
    prim_el_rep_ey = {}
    for i in range(len(ey)):
        prim_el_rep_ey[i] = prim_el_rep(ey[i],d,prim_el,sx)
    deg_ey = len(ey) - 1     
    qy = 0 
    ry = 0
    
    while(deg_ny>=deg_ey):
        diff = deg_ny - deg_ey
        temp_el = prim_el_rep_ny[deg_ny]
        print("prim_el to the power:",temp_el)
        print("y to the power:",diff)
#         qx,rx = div(prim_el**temp_el,sx)
        qy = qy + (prim_el**temp_el)*y**diff
        for i in range(len(ey)):
            prim_el_rep_ny[diff+i] = prim_el_rep_sub(prim_el_rep_ny[diff+i],prim_el_rep_mul(temp_el,prim_el_rep_ey[i],d),d,prim_el,sx)
        prim_el_rep_ny[deg_ny] = "z"
        deg_ny =  deg_ny_find(prim_el_rep_ny)
    
    ry = pol_from_prim_el_rep(prim_el_rep_ny,y)
    
    return qy,ry
```

## Decoding message polynomial from received vector 

```python
# R_cdrwd[i] * E(eval_points[i]) = N(eval_points[i]), i =1,2...n
# n equations in ((t_err)+1)+((t_err+k-1)+1) variables 
# E(y) is t_err degree polynomial so we fix e_t_err=1
# n equations in at most t_err + (t_err + k) <= n variables 
# Coeff_matrix * variables_vector = constant_vector
# Find the solution using gaussian elimination 

# Over F2: -1 = +1 

# First t_err columns of Coeff_matrix correspond to coefficients of E(y)
# Next t_err+k columns of Coeff_matrix correspond to coefficients of N(y)
# i^th row of Coeff_matrix would look like:
# yi,yi*alpha_i,yi*(alpha_i**2),...,yi*(alpha_i**(t_err-1)),1,alpha_i,alpha_i**2,...,alpha_i**(k+t_err-1)

# variables_vector looks like: e_0,e_1,...,e_t_err-1,n_0,n_1,...,n_k+t_err-1

# i^th element of constant_vector would look like: yi*(alpha_i**t_err)

Coeff_matrix = {}
for i in range (0,n):
    Coeff_matrix[i]= {}
    for j in range (0,t_err):
        qx,rx = div(r_cdwrd[i]*(F2_4[eval_points[i]]**j),sx)
        Coeff_matrix[i][j] = mod2_pol(rx)
    for j in range (t_err,k+2*t_err):
        qx,rx = div(F2_4[eval_points[i]]**(j-t_err),sx)
        Coeff_matrix[i][j] = mod2_pol(rx)

constant_vector = {}
for i in range (0,n):
    qx,rx = div(r_cdwrd[i]*(F2_4[eval_points[i]]**t_err),sx)
    constant_vector[i] = mod2_pol(rx)
    
aug_matrix = {}
for i in range(0,n):
    aug_matrix[i] = {}
    for j in range(0,n):
        aug_matrix[i][j]=Coeff_matrix[i][j]
    aug_matrix[i][n] = constant_vector[i]
```

```python
def sp_gaus_elim(aug_matrix_prim_rep,n,cmcolnum,d,prim_el,sx):
# Apply modified(modified arithmetic operations) gaussian elimination algorithm on aug_matrix_prim_rep

# To convert it into row echelon form
# Step 1: Pivot the matrix 
# Note: As the number of variables are more than the number of equations pivot might not be at i,i 
# Step 2: To get the matrix in row echelon form, repeat the pivot

# Find the first non-zero entry in the pivot column(!=pivot_row) of the matrix 
    pvt_col = 0
    pvt_row = 0
    while (pvt_row < n and pvt_col < cmcolnum ):
#         print(pd.DataFrame(aug_matrix_prim_rep).transpose())
        print("pvt_col:",pvt_col)
        row = "NA"
        skip_flag = 0

        for i in range (pvt_row,n):
            if (aug_matrix_prim_rep[i][pvt_col] != "z"):
                row = i
                break
        print("row:",row)
        if (row=="NA"):
            skip_flag = 1
            pvt_col = pvt_col+1


        if(skip_flag==0):    
            print("pvt_row:", pvt_row)
            # Interchange rows, moving the row to the pivot row        
            temp = {}
            temp = aug_matrix_prim_rep[pvt_row]
            aug_matrix_prim_rep[pvt_row] =aug_matrix_prim_rep[row]
            aug_matrix_prim_rep[row] = temp 


            # Multiply each element in the pivot row by the inverse of the pivot, so the pivot equals 1
            pvt_el_inv = prim_el_rep_inverse(aug_matrix_prim_rep[pvt_row][pvt_col])
            for i in range (0,cmcolnum+1):
                aug_matrix_prim_rep[pvt_row][i] = prim_el_rep_mul(pvt_el_inv,aug_matrix_prim_rep[pvt_row][i],d)

            # Add multiples of the pivot row to each of the lower rows, 
            # so every element in the pivot column of the lower rows equals 0.

            for j in range(pvt_row+1, n):
                if(aug_matrix_prim_rep[j][pvt_col]=='z'):
                    continue
                else:
                    ratio = prim_el_rep_div(aug_matrix_prim_rep[j][pvt_col],aug_matrix_prim_rep[pvt_row][pvt_col],d)

                for l in range(0,cmcolnum+1):
                    p_2 = prim_el_rep_mul(ratio,aug_matrix_prim_rep[pvt_row][l],d)
                    aug_matrix_prim_rep[j][l]= prim_el_rep_sub(aug_matrix_prim_rep[j][l],p_2,d,prim_el,sx)
            pvt_col = pvt_col + 1
            pvt_row = pvt_row + 1
    return pvt_row-1,pvt_col-1

def sp_rre_form(aug_matrix_prim_rep,n,cmcolnum,last_pvt_row,last_pvt_col):
# Convert the aug_matrix_prim_rep from row echelon form to row reduced echelon form 
    pvt_col = last_pvt_col
    
    for pvt_row in range(last_pvt_row,0,-1):
        
        for u in range (0,cmcolnum):
            if(aug_matrix_prim_rep[pvt_row][u]!="z"):
                pvt_col = u
                break
                
        for j in range(pvt_row-1,-1,-1):
            if(aug_matrix_prim_rep[j][pvt_col]=='z'):
                continue
            else:
                ratio = prim_el_rep_div(aug_matrix_prim_rep[j][pvt_col],aug_matrix_prim_rep[pvt_row][pvt_col],d)

            for l in range(0,cmcolnum):
                p_2 = prim_el_rep_mul(ratio,aug_matrix_prim_rep[pvt_row][l],d)
                aug_matrix_prim_rep[j][l]= prim_el_rep_sub(aug_matrix_prim_rep[j][l],p_2,d,prim_el,sx)
                    
```

### STEP I: Interpolation step 

```python
XY_pows = {}
n_terms = 0
# i + (k-1)*j <= D
l_wt = int(np.floor(D/(k-1)))
for i in range(0,l_wt+1):
    XY_pows[i]={}
    for j in range(0,D+1):
        if ((k-1)*i + j <= D):
            XY_pows[i][j]=j
            n_terms = n_terms + 1
```

```python
# Q(X,Y) = sum(i from 0 to l_wt) sum(j from 0 to D-(k-1)*i) q_ij*X^j*Y^i

# Q(eval_points[l],r_cdrwd[l])=0 for l =1,2...n
# n equations in n_terms variables

# Coeff_matrix * variables_vector = constant_vector

cmcolnum = n_terms

Coeff_matrix = {}
for l in range (0,n):
    cnt = 0
    Coeff_matrix[l]= {}
    for i in range (0,l_wt+1):
        for j in range (0,len(XY_pows[i])):
            qx,rx = div((F2_d[eval_points[l]]**XY_pows[i][j])*(r_cdwrd[l]**i),sx)
            Coeff_matrix[l][cnt] = mod2_pol(rx)
            cnt = cnt + 1 

constant_vector = {}
for i in range (0,cmcolnum):
    constant_vector[i] = 0
    
aug_matrix = {}
for i in range(0,n):
    aug_matrix[i] = {}
    for j in range(0,cmcolnum):
        aug_matrix[i][j]= Coeff_matrix[i][j]
    aug_matrix[i][cmcolnum] = constant_vector[i]
```
